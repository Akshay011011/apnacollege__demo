```python
import os
import numpy as np

def load_dataset(root_dir, split):
    in_dir = os.path.join(root_dir, split)
    data_paths = [os.path.join(in_dir, f) for f in os.listdir(in_dir) if os.path.isfile(os.path.join(in_dir, f))]
    data_and_seg_arr = [np.load(ii, mmap_mode='r') for ii in data_paths]
    pids = [ii.split('/')[-1].split('.')[0] for ii in data_paths]
    data = {}
    for ix, pid in enumerate(pids):
        data[pid] = {'data': data_and_seg_arr[ix][..., 0], 'seg': data_and_seg_arr[ix][..., 1], 'pid': pid}
    return data

# Example usage:
root_directory = '/kaggle/input/mnist-dataset'
split_type = 'train'
loaded_data = load_dataset(root_directory, split_type)
```

```python
!pip install python-mnist

from mnist import MNIST
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Path to the directory containing the MNIST files
mnist_path = '/kaggle/input/mnist-dataset'

# Initialize the MNIST reader
mndata = MNIST(mnist_path)

# Load the training and test data
X_train, y_train = mndata.load_training()
X_test, y_test = mndata.load_testing()

Convert to numpy arrays
X_train = np.array(X_train)
y_train = np.array(y_train)
X_test = np.array(X_test)
y_test = np.array(y_test)

# Print shapes to verify
print(f"Training features shape: {X_train.shape}")
print(f"Training labels shape: {y_train.shape}")
print(f"Testing features shape: {X_test.shape}")
print(f"Testing labels shape: {y_test.shape}")

# Function to preprocess MNIST images
def preprocess_mnist_images(images):
    processed_images = []
    for image in images:
        reshaped_img = image.reshape(28, 28)
        normalized_img = cv2.normalize(reshaped_img, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
        processed_images.append(normalized_img)
    return np.array(processed_images)

# Preprocess the images
X_train_processed = preprocess_mnist_images(X_train)
X_test_processed = preprocess_mnist_images(X_test)

# Print the shapes to verify
print(f"Processed training features shape: {X_train_processed.shape}")
print(f"Processed testing features shape: {X_test_processed.shape}")

# Display a sample image and its label
plt.imshow(X_train_processed[0], cmap='gray')
plt.title(f'Label: {y_train[0]}')
plt.axis('off')
plt.show()
```

```python
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Load the uploaded file
# file_path = '/kaggle/input/mnist-dataset'

# Read the file using pandas
data = pd.read_csv('/kaggle/input/mnsitdataset/train-images-idx3-ubyte')

# Display the first few rows to understand the structure
print(data.head())

# Separate features and labels
X = data.drop('label', axis=1).values
y = data['label'].values

# Print shapes to verify
print(f"Features shape: {X.shape}")
print(f"Labels shape: {y.shape}")

# Function to preprocess MNIST images
def preprocess_mnist_images(images):
    processed_images = []
    for image in images:
        reshaped_img = image.reshape(28, 28)
        normalized_img = cv2.normalize(reshaped_img, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
        processed_images.append(normalized_img)
    return np.array(processed_images)

# Preprocess the images
X_processed = preprocess_mnist_images(X)

# Print the shape to verify
print(f"Processed features shape: {X_processed.shape}")

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42)

# Print the shapes to verify
print(f"Training features shape: {X_train.shape}")
print(f"Training labels shape: {y_train.shape}")
print(f"Testing features shape: {X_test.shape}")
print(f"Testing labels shape: {y_test.shape}")

# Display a sample image and its label
plt.imshow(X_train[0], cmap='gray')
plt.title(f'Label: {y_train[0]}')
plt.axis('off')
plt.show()
```
